```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate grid for contour plot
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(-2, 2, 400)
X1, X2 = np.meshgrid(x1, x2)

Z = -1 + X1**2 + X2**2
H = sigmoid(Z)

# Generate sample points
np.random.seed(0)
n_points = 200
x_points = np.random.uniform(-2, 2, (n_points, 2))
z_points = -1 + x_points[:, 0]**2 + x_points[:, 1]**2
y_points = (sigmoid(z_points) >= 0.5).astype(int)  # label: 1 if inside, 0 if outside

# Plotting
plt.figure(figsize=(6, 6))
plt.contour(X1, X2, H, levels=[0.5], colors='red', linewidths=2, linestyles='--', label='Decision Boundary')
plt.scatter(x_points[y_points == 1][:, 0], x_points[y_points == 1][:, 1], c='blue', label='Class 1 (inside)', alpha=0.7)
plt.scatter(x_points[y_points == 0][:, 0], x_points[y_points == 0][:, 1], c='green', label='Class 0 (outside)', alpha=0.7)

plt.title(r'Decision Boundary: $h_\theta(x) = \sigma(-1 + x_1^2 + x_2^2)$')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Parameters for h(x) = ( + x)
theta0 = -1.0
theta1 = 1.0

# Create mesh grid (x2 is dummy, just for plotting in 2D)
x1 = np.linspace(-5, 5, 400)
x2 = np.linspace(-5, 5, 400)
X1, X2 = np.meshgrid(x1, x2)

# Compute h_theta(x) = ( + x) -- x2 is ignored
Z = theta0 + theta1 * X1
H = sigmoid(Z)

# Decision boundary:  + x = 0  x = - / 
x1_boundary = -theta0 / theta1

# Generate data points
np.random.seed(0)
X = np.random.uniform(-5, 5, (300, 2))
z_points = theta0 + theta1 * X[:, 0]
y = (sigmoid(z_points) >= 0.5).astype(int)

# Plot
plt.figure(figsize=(6, 6))
plt.contour(X1, X2, H, levels=[0.5], colors='red', linestyles='--', linewidths=2)
plt.axvline(x=x1_boundary, color='red', linestyle='--', label='Decision Boundary')

plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], c='blue', edgecolor='k', label='Class 1')
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], c='green', edgecolor='k', label='Class 0')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$ (ignored)')
plt.title(r'Decision Boundary: $h_\theta(x) = \sigma(\theta_0 + \theta_1 x_1)$')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate synthetic data
np.random.seed(42)
x = np.linspace(-2, 2, 100)
# Actual target y based on some arbitrary polynomial relationship
z_true = 0.5*x**2 - 1.2*x**1 + 0.8*x
y_actual = (sigmoid(z_true) > 0.5).astype(float)

# Create a grid of weights for w2 and w3 to plot the surface
w2_range = np.linspace(-5, 5, 50)
w3_range = np.linspace(-5, 5, 50)
W2, W3 = np.meshgrid(w2_range, w3_range)

# Fixed parameters
w1 = 0.8
b = 0

# Calculate MSE over the grid
mse_surface = np.zeros(W2.shape)

for i in range(len(w3_range)):
    for j in range(len(w2_range)):
        # Calculate logit with degree 3 polynomial
        z_pred = W3[i,j]*x**2 + W2[i,j]*x**1 + w1*x + b
        y_pred = sigmoid(z_pred)
        mse_surface[i,j] = np.mean((y_actual - y_pred)**2)

# Plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(W2, W3, mse_surface, cmap='magma', edgecolor='none', alpha=0.9)

ax.set_title('Non-Convex MSE Loss Surface (3rd Degree Polynomial)', fontsize=14)
ax.set_xlabel('Weight w2', fontsize=12)
ax.set_ylabel('Weight w3', fontsize=12)
ax.set_zlabel('MSE Loss', fontsize=12)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

# 1. Generate synthetic data
np.random.seed(42)
x = np.linspace(-1, 1, 50)
# Define a target where the 'true' w3 coefficient is 5
z_true = 0.5*x**2 - 1.2*x**1 + 0.8*x
y_actual = (sigmoid(z_true) > 0.5).astype(float)

# 2. Fix other parameters and vary only w3
w2_fixed = -2
w1_fixed = 0
b_fixed = 0

w3_values = np.linspace(-20, 20, 500)
mse_values = []

for w3 in w3_values:
    z_pred = w3 * x**2 + w2_fixed * x**1 + w1_fixed * x + b_fixed
    y_pred = sigmoid(z_pred)
    mse_values.append(mse_loss(y_actual, y_pred))

# 3. Create the 2D Plot
plt.figure(figsize=(10, 6))
plt.plot(w3_values, mse_values, color='red', linewidth=2, label='MSE Loss')

plt.title('2D Slice of MSE Loss Surface (Varying $w_3$ in a 3rd Degree Polynomial)', fontsize=14)
plt.xlabel('Weight $w_3$ (Coefficient of $x^3$)', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.show()
```
