# Simple graph

Draw simple graph with manual layout.


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(1, 5)
G.add_edge(2, 3)
G.add_edge(3, 4)
G.add_edge(4, 5)

# explicitly set positions
pos = {1: (0, 0), 2: (-2, 0.3), 3: (2, 0.17), 4: (4, 0.255), 5: (5, 0.03)}

options = {
    "font_size": 16,
    "node_size": 1000,
    "node_color": "yellow",
    "edgecolors": "blue",
    "linewidths": 2,
    "width": 1,
}
nx.draw_networkx(G, pos)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

A directed graph



```python
G = nx.DiGraph([(0, 3), (1, 3), (2, 4), (3, 5), (3, 6), (4, 6), (5, 6)])

# group nodes by column
left_nodes = [0, 1, 2]
middle_nodes = [3, 4]
right_nodes = [5, 6]

# set the position according to column (x-coord)
pos = {n: (0, i) for i, n in enumerate(left_nodes)}
pos.update({n: (1, i + 0.5) for i, n in enumerate(middle_nodes)})
pos.update({n: (2, i + 0.5) for i, n in enumerate(right_nodes)})

nx.draw_networkx(G, pos, **options)

# Set margins for the axes so that nodes aren't clipped
ax = plt.gca()
ax.margins(0.20)
plt.axis("off")
plt.show()
```

```python
# Importing the dataset
G = nx.karate_club_graph()

# Statistics of the graph
print(f"Number of nodes: {len(G.nodes())}")
print(f"Number of edges: {len(G.edges())}")
average_clustering = nx.average_clustering(G)
print(f"Average Clustering Coefficient: {average_clustering}")
diameter = nx.diameter(G)
print(f"Graph Diameter: {diameter}")
density = nx.density(G)
print(f"Graph Density: {density}")
# Number of nodes: 34
# Number of edges: 78
# Average Clustering Coefficient: 0.57
# Graph Diameter: 5
# Graph Density: 0.14

# Plotting graph with Matplotlib and Networkx
plt.figure(figsize=(7,5))
nx.draw_networkx(G, with_labels=True, font_weight='bold')
plt.show()
```

```python
# Compute degree centrality
degree_centrality = nx.degree_centrality(G)

# Sorting and printing the degree centrality
sorted_degree_centrality = sorted(degree_centrality.items(), key=lambda x:x[1], reverse=True)
print("Degree Centrality:", sorted_degree_centrality)
```

```python
# Compute betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Sorting and printing the closenees centrality
sorted_betweenness_centrality = sorted(betweenness_centrality.items(), key=lambda x:x[1], reverse=True)
print("Betweenness Centrality:", sorted_betweenness_centrality)
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# 1️⃣ Load Excel file
df = pd.read_csv('sna3.csv', index_col=0)

# 2️⃣ Extract node numbers as string (optional: convert to string or keep integer)
nodes = df.index.tolist()

# 3️⃣ Initialize undirected graph
G = nx.Graph()

# 4️⃣ Add nodes
G.add_nodes_from(nodes)

# 5️⃣ Add edges where matrix == 1
for i, source in enumerate(nodes):
    for j, target in enumerate(nodes):
        if df.iloc[i, j] == 1:
            # Avoid duplicate edges — add only if source < target (undirected graph)
            if source < target:
                G.add_edge(source, target)

# 6️⃣ Draw the graph
plt.figure(figsize=(14, 12))
pos = nx.circular_layout(G, scale=2)

nx.draw_networkx_nodes(G, pos, node_color='lightblue', node_size=800)
nx.draw_networkx_edges(G, pos, edge_color='gray', width=2)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

plt.title("Graph of Person Relationships from Matrix", fontsize=16)
plt.axis('off')
plt.show()
```

```python
# Compute degree centrality
degree_centrality = nx.degree_centrality(G)

# Sorting and printing the degree centrality
sorted_degree_centrality = sorted(degree_centrality.items(), key=lambda x:x[1], reverse=True)
print("Degree Centrality:", sorted_degree_centrality)
```

```python
# Compute closeness centrality
closeness_centrality = nx.closeness_centrality(G)

# Sorting and printing the closenees centrality
sorted_closeness_centrality = sorted(closeness_centrality.items(), key=lambda x:x[1], reverse=True)
print("Closeness Centrality:", sorted_closeness_centrality)
```

```python
# Compute betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Sorting and printing the closenees centrality
sorted_betweenness_centrality = sorted(betweenness_centrality.items(), key=lambda x:x[1], reverse=True)
print("Betweenness Centrality:", sorted_betweenness_centrality)
```

```python
# community detection with Louvain Community Detection Algorithm
communities = nx.community.louvain_communities(G)

# print all the communities
for i, community in enumerate(communities):
    print(f"{i+1}° Community : {community}")
```
