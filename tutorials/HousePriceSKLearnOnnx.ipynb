```python
import pandas as pd
import numpy as np

data = pd.read_csv('https://raw.githubusercontent.com/ekaratnida/Applied-machine-learning/master/Week04-workshop-1/data.txt')
data.head()
```

```python
from sklearn.linear_model import LinearRegression
X,y = data.values[:,0:2], data.values[:,2]
print(X)
print(y)
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
print(f"Predict the house price of the house which has 3 rooms and 2000 area = {reg.predict([[2000,3]])[0]}")
```

```python
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType

initial_type = [
    ('input_area_and_rooms', FloatTensorType([None, 2]))
]
converted_model = convert_sklearn( reg, initial_types = initial_type)
with open("house_price_model.onnx", "wb") as f:
  f.write(converted_model.SerializeToString())
```
