```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate grid for contour plot
x1 = np.linspace(-2, 2, 400)
x2 = np.linspace(-2, 2, 400)
X1, X2 = np.meshgrid(x1, x2)

Z = -1 + X1**2 + X2**2
H = sigmoid(Z)

# Generate sample points
np.random.seed(0)
n_points = 200
x_points = np.random.uniform(-2, 2, (n_points, 2))
z_points = -1 + x_points[:, 0]**2 + x_points[:, 1]**2
y_points = (sigmoid(z_points) >= 0.5).astype(int)  # label: 1 if inside, 0 if outside

# Plotting
plt.figure(figsize=(6, 6))
plt.contour(X1, X2, H, levels=[0.5], colors='red', linewidths=2, linestyles='--', label='Decision Boundary')
plt.scatter(x_points[y_points == 1][:, 0], x_points[y_points == 1][:, 1], c='blue', label='Class 1 (inside)', alpha=0.7)
plt.scatter(x_points[y_points == 0][:, 0], x_points[y_points == 0][:, 1], c='green', label='Class 0 (outside)', alpha=0.7)

plt.title(r'Decision Boundary: $h_\theta(x) = \sigma(-1 + x_1^2 + x_2^2)$')
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Parameters for h(x) = σ(θ₀ + θ₁x₁ + θ₂x₂)
theta0 = 0.5
theta1 = 1
theta2 = -1

# Generate grid of points
x1 = np.linspace(-5, 5, 400)
x2 = np.linspace(-5, 5, 400)
X1, X2 = np.meshgrid(x1, x2)

# Compute h_theta(x) = σ(θ₀ + θ₁x₁ + θ₂x₂)
Z = theta0 + theta1 * X1 + theta2 * X2
H = sigmoid(Z)

# Decision boundary where h = 0.5 ⇔ z = 0 ⇔ θ₀ + θ₁x₁ + θ₂x₂ = 0
# This is a straight line: x₂ = -(θ₀ + θ₁x₁)/θ₂
x1_line = np.linspace(-5, 5, 100)
x2_line = -(theta0 + theta1 * x1_line) / theta2

# Sample points to plot (optional)
np.random.seed(0)
X = np.random.uniform(-5, 5, (300, 2))
z_points = theta0 + theta1 * X[:, 0] + theta2 * X[:, 1]
y = (sigmoid(z_points) >= 0.5).astype(int)

# Plot
plt.figure(figsize=(6, 6))
plt.contour(X1, X2, H, levels=[0.5], colors='red', linestyles='--', linewidths=2)
plt.plot(x1_line, x2_line, 'r--', label='Decision Boundary')

plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], c='blue', edgecolor='k', label='Class 1')
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], c='green', edgecolor='k', label='Class 0')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.title(r'Decision Boundary: $h_\theta(x) = \sigma(\theta_0 + \theta_1 x_1 + \theta_2 x_2)$')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Parameters for h(x) = σ(θ₀ + θ₁x₁)
theta0 = -1.0
theta1 = 1.0

# Create mesh grid (x2 is dummy, just for plotting in 2D)
x1 = np.linspace(-5, 5, 400)
x2 = np.linspace(-5, 5, 400)
X1, X2 = np.meshgrid(x1, x2)

# Compute h_theta(x) = σ(θ₀ + θ₁x₁) -- x2 is ignored
Z = theta0 + theta1 * X1
H = sigmoid(Z)

# Decision boundary: θ₀ + θ₁x₁ = 0 ⇒ x₁ = -θ₀ / θ₁
x1_boundary = -theta0 / theta1

# Generate data points
np.random.seed(0)
X = np.random.uniform(-5, 5, (300, 2))
z_points = theta0 + theta1 * X[:, 0]
y = (sigmoid(z_points) >= 0.5).astype(int)

# Plot
plt.figure(figsize=(6, 6))
plt.contour(X1, X2, H, levels=[0.5], colors='red', linestyles='--', linewidths=2)
plt.axvline(x=x1_boundary, color='red', linestyle='--', label='Decision Boundary')

plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], c='blue', edgecolor='k', label='Class 1')
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], c='green', edgecolor='k', label='Class 0')

plt.xlabel('$x_1$')
plt.ylabel('$x_2$ (ignored)')
plt.title(r'Decision Boundary: $h_\theta(x) = \sigma(\theta_0 + \theta_1 x_1)$')
plt.legend()
plt.grid(True)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
