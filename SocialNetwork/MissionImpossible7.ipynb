```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.karate_club_graph()
pos = nx.spring_layout(G)

labels = nx.get_node_attributes(G, 'club')  # 'Mr. Hi' or 'Officer'

colors = ['lightblue' if labels[n] == 'Mr. Hi' else 'orange' for n in G.nodes]

plt.figure(figsize=(10, 8))
nx.draw(G, pos, with_labels=True, node_color=colors, edge_color='gray')
plt.title("Zachary's Karate Club Network")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
G = nx.DiGraph()

# Add nodes
entities = [
    "Ethan Hunt", "Ilsa Faust", "Gabriel", "Grace", "Benji Dunn", "Luther Stickell",
    "The Entity", "White Widow", "CIA", "IMF", "Key", "Russian Submarine"
]
G.add_nodes_from(entities)

# Add edges (relationships)
edges = [
    ("Ethan Hunt", "Ilsa Faust", "Ally"),
    ("Ethan Hunt", "Gabriel", "Nemesis"),
    ("Ethan Hunt", "Grace", "Protector"),
    ("Ethan Hunt", "The Entity", "Opposes"),
    ("Gabriel", "The Entity", "Agent"),
    ("Ethan Hunt", "Benji Dunn", "Friend"),
    ("Ethan Hunt", "Luther Stickell", "Friend"),
    ("Benji Dunn", "IMF", "Member"),
    ("Luther Stickell", "IMF", "Member"),
    ("Ethan Hunt", "IMF", "Member"),
    ("IMF", "CIA", "Tension"),
    ("Grace", "White Widow", "Rival"),
    ("Key", "Russian Submarine", "Opens Access"),
    ("The Entity", "Russian Submarine", "Controls"),
]

for src, dst, label in edges:
    G.add_edge(src, dst, label=label)

# Draw the graph
pos = nx.spring_layout(G, k=0.8)
edge_labels = nx.get_edge_attributes(G, 'label')

plt.figure(figsize=(14, 10))
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=3000, font_size=10)
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')
plt.title("Mission: Impossible  Dead Reckoning (Graph of Entities)", fontsize=16)
plt.show()
```
