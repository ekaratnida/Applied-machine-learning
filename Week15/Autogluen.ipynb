```python
from autogluon.tabular import TabularPredictor
my_models = {
    'GBM': {},       # LightGBM (Great for speed)
    'RF': {},        # Random Forest (Robust)
    'XGB': {},       # XGBoost (Industry standard)
    'CAT': {},       # CatBoost (Excellent for categorical data)
}

#predictor = TabularPredictor(label='target').fit(
#    train_data,
#    hyperparameters=my_models
#)

predictor = TabularPredictor(label="Survived").fit(
    "train_data.csv",
    hyperparameters=my_models,
    presets='best_quality'
    )
predictions = predictor.predict("test_data.csv")
```

```python
predictor.evaluate("test_data.csv")
```
