```python
# prompt: please write python code for all math functions of numpy on d variable.

d = [
[19, 87, 63],
[20, 71, 54 ]
]
d_array = np.array(d)

# Basic arithmetic operations
print("Addition (scalar):", np.add(d_array, 5))
print("Subtraction (scalar):", np.subtract(d_array, 5))
print("Multiplication (scalar):", np.multiply(d_array, 5))
print("Division (scalar):", np.divide(d_array, 5))
print("Power (scalar):", np.power(d_array, 2))
print("Modulus (scalar):", np.mod(d_array, 3))

print("Addition (array):", np.add(d_array, d_array))
print("Subtraction (array):", np.subtract(d_array, d_array))
print("Multiplication (array):", np.multiply(d_array, d_array))
print("Division (array):", np.divide(d_array, d_array))


# Trigonometric functions
print("Sine:", np.sin(d_array))
print("Cosine:", np.cos(d_array))
print("Tangent:", np.tan(d_array))
print("Arc Sine:", np.arcsin(d_array / np.max(d_array)))  # Normalize to avoid errors
print("Arc Cosine:", np.arccos(d_array / np.max(d_array)))
print("Arc Tangent:", np.arctan(d_array))


# Exponential and logarithmic functions
print("Exponential:", np.exp(d_array))
print("Logarithm (base e):", np.log(d_array))
print("Logarithm (base 10):", np.log10(d_array))


# Other mathematical functions
print("Square Root:", np.sqrt(d_array))
print("Absolute Value:", np.abs(d_array))
print("Ceiling:", np.ceil(d_array))
print("Floor:", np.floor(d_array))
print("Round:", np.round(d_array))

print("Maximum element:", np.max(d_array))
print("Minimum element:", np.min(d_array))
print("Sum of all elements:", np.sum(d_array))
print("Mean of all elements:", np.mean(d_array))
print("Standard deviation:", np.std(d_array))
print("Variance:", np.var(d_array))

print("Dot product:", np.dot(d_array, d_array.T))  # Note: Dot product requires compatible shapes
```

```python
# prompt: Please plot bar chart of d variable

import matplotlib.pyplot as plt

d = [
[19, 87, 63],
[20, 71, 54 ]
]

# Assuming d is a list of lists representing data for the bar chart
# Each inner list represents a set of values for a different bar

# Convert the list of lists to a numpy array
d_array = np.array(d)

# Create bar chart using matplotlib
plt.figure(figsize=(8, 6))

# Create bar plots for each inner list in d
for i in range(d_array.shape[0]):
  plt.bar(np.arange(d_array.shape[1]) + i * 0.25, d_array[i], width=0.25, label=f"Bar {i+1}")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Chart")
plt.legend()
plt.show()
```

```python
# prompt: How about donut chart on d variable

import matplotlib.pyplot as plt

d = [
[19, 87, 63],
[20, 71, 54 ]
]

# Assuming d is a list of lists representing data for the donut chart
# Each inner list represents a set of values for a different category

# Flatten the list of lists into a single list
data = [item for sublist in d for item in sublist]

# Create labels for each segment of the donut chart
labels = [f"Segment {i+1}" for i in range(len(data))]

# Create the donut chart
plt.figure(figsize=(8, 6))
plt.pie(data, labels=labels, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))

# Add a white circle in the center to create a donut shape
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Donut Chart")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# prompt: ช่วย plot chart ที่ดูง่ายๆ กับข้อมูล df_num.corr() ให้หน่อยครับ โดย chart สามารถ Interactive กับผู้ใช้ได้ด้วยครับ เช่น plotly ประมาณนั้นครับ

import plotly.express as px

# Calculate the correlation matrix
correlation_matrix = df_num.corr()

# Create a heatmap using Plotly
fig = px.imshow(correlation_matrix,
                labels=dict(x="Features", y="Features", color="Correlation"),
                x=correlation_matrix.columns,
                y=correlation_matrix.columns,
                color_continuous_scale='RdBu',  # Use a diverging color scale
                title="Correlation Matrix")

# Customize the layout
fig.update_layout(
    width=800,
    height=600,
    xaxis_tickfont_size=10,
    yaxis_tickfont_size=10
)

# Show the interactive plot
fig.show()
```
